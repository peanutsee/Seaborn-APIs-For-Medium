## Relational Plots

#### relplot
Figure-level interface for drawing relational plots onto a FacetGrid. Shows the relationship between two continous data col, hue can be defined to categorise by categorical data col(s).

https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot

```python
sns.set(style= 'ticks')
g = sns.relplot(x= 'total_bill', y= 'tip', hue= 'day', data= df)
g.savefig('relplot 1.png')
```

```python
print('Continous Data Col:', 'tips & total_bill')
print('Categorical Data Col:', 'day')
```

```python
g = sns.relplot(x= 'total_bill', y= 'tip', hue= 'day', col= 'time', data= df)
g.savefig('relplot 2.png')
```

```python
print('Continous Data Col:', 'tips & total_bill')
print('Categorical Data Col:', 'day & time')
```

```python
g = sns.relplot(x= 'total_bill', y= 'tip', hue= 'day', col= 'sex', row= 'time', data= df)
g.savefig('relplot 3.png')
```

```python
print('Continous Data Col:', 'tips & total_bill')
print('Categorical Data Col:', 'day, time & row') 
```

```python
g = sns.relplot(x="total_bill", y="tip", hue="time", size="size",
                palette=["b", "r"], sizes=(10, 100), col= 'time', data=df)
```

```python
g = sns.relplot(x= 'total_bill', y= 'tip', hue= 'time', col= 'day', kind= 'line', data= df)
g.savefig('relplot 5.png')
```

To summarise:

Compare the relationship between 2 continous plots, hue, style, col and rows can be defined to categorious the continous data according to their corresponding categorical data.

1. boxplot() (kind= 'box')

```python
# Overall plots
g = sns.PairGrid(iris, hue= 'species')
g = g.map(sns.scatterplot)
```

```python
# Diag Plots and off_diag
g = sns.PairGrid(iris, hue= 'species')
g = g.map_diag(plt.hist)
g = g.map_offdiag(plt.scatter)
```

```python
# Diag Plots and off_diag
g = sns.PairGrid(iris, hue= 'species')
g = g.map_diag(plt.hist, histtype= 'step', linewidth= 3)
g = g.map_offdiag(sns.scatterplot)
```

```python
# Subset Plotting
g = sns.PairGrid(iris, vars= ['sepal_length', 'sepal_width'], hue= 'species')
g = g.map_diag(plt.hist, histtype= 'step', linewidth= 1.5)
g = g.map_offdiag(sns.scatterplot)
```

```python
# Ultimate Customisation
g = sns.PairGrid(iris, hue= 'species', palette= 'Set2')
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.distplot)
g = g.map_upper(sns.scatterplot)
```

```python
# JointGrid() -Grid for drawing a bivariate plot with marinal univariate plots
tips = sns.load_dataset('tips')
g = sns.JointGrid(x= 'total_bill', y= 'tip', data= tips)
```

```python
# Doesn't provide much... customs
g = sns.JointGrid(x= 'total_bill', y= 'tip', data= tips)
g = g.plot(sns.regplot, sns.distplot)
```

```python
# Customising joint and marginals
g = sns.JointGrid(x= 'total_bill', y= 'tip', data= tips)
g = g.plot_joint(sns.scatterplot, color= 'm')
g = g.plot_marginals(sns.distplot, color= '0.1')
```

```python
# Customising joint and individual marginals
g = sns.JointGrid(x= 'total_bill', y= 'tip', data= tips)
g = g.plot_joint(sns.scatterplot, color= 'm')
_ = g.ax_marg_x.hist(tips['total_bill'], color= 'b', alpha= 0.6)
_ = g.ax_marg_y.hist(tips['tip'], color= 'r', alpha= 0.6, orientation= 'horizontal')
```

```python
# Removing space between joint and marginals
g = sns.JointGrid(x="total_bill", y="tip", data=tips, space=0)
g = g.plot_joint(sns.kdeplot, cmap="Blues_d")
g = g.plot_marginals(sns.kdeplot, shade=True)
```
